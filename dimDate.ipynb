```
# https://www.bluegranite.com/blog/generate-a-calendar-dimension-in-spark
from pyspark.sql.functions import explode, sequence, to_date

beginDate = '1940-01-01'
endDate = '2023-12-31'

(
  spark.sql(f"select explode(sequence(to_date('{beginDate}'), to_date('{endDate}'), interval 1 day)) as calendarDate")
    .createOrReplaceTempView('dates')
)
```

```
%sql
create or replace table Date
using delta
location '/delta/Date'
as select
  calendarDate AS Date,
  year(calendarDate) AS Year,
  quarter(calendarDate) as QuarterOfYear,
  month(calendarDate) as Month,
  dayofmonth(calendarDate) as Day,
  dayofyear(calendarDate) as DayOfYear,
  dayofweek(calendarDate) - 1 AS DayOfWeek,
  case
    when dayofweek(calendarDate) < 6 then False
    else True
  end as IsWeekend

from
  dates
order by
  Date
```
